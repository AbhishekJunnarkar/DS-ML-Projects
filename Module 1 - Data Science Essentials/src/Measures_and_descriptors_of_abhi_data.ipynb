```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df_ = pd.read_excel('data_abhi1.xlsx')  #importing excel sheet
df_.head()
```

```python
# Visualizing the dataset
## Plot histogram for Age
### Use Case: Show the frequency of a numeric variable (e.g., Age, Score).
plt.hist(df_["Age"], bins=10, color="skyblue", edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution Histogram (Distribution of Age among the population)")
plt.show()
```

```python
## Box Plot (Outliers and Spread)
### Use Case: Detect outliers in Height or Weight.
sns.boxplot(x=df_["Height (cm)"])
plt.title("Box Plot of Heights")
plt.show()
```

```python
## Scatter Plot (Correlation Between Two Variables)
### Use Case: Show the relationship between Height and Weight.
plt.scatter(df_["Height (cm)"], df_["Weight (kg)"], color="green")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Height vs Weight")
plt.show()
```

```python
## Bar Chart (Comparison of Categories)
### Use Case: Compare average scores by age group.
df_["Age Group"] = pd.cut(df_["Age"], bins=[18, 30, 40, 50, 60], labels=["18-30", "30-40", "40-50", "50-60"])
avg_scores = df_.groupby("Age Group")["Score"].mean()

avg_scores.plot(kind="bar", color="purple")
plt.xlabel("Age Group")
plt.ylabel("Average Score")
plt.title("Average Score by Age Group")
plt.show()
```

```python
## Pie Chart (Percentage Distribution)
### Use Case: Show the proportion of people in different age groups.
age_counts = df_["Age Group"].value_counts()
age_counts.plot(kind="pie", autopct="%1.1f%%", colors=["red", "blue", "green", "yellow"])
plt.title("Age Group Distribution")
plt.ylabel("")
plt.show()
```

```python
## Line Plot (Trends Over Time)
### Use Case: If your data has a time element, plot trends.
df_["ID"].plot(kind="line", y="Score", color="orange")
plt.xlabel("ID")
plt.ylabel("Score")
plt.title("Score Trend")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
```

```python


# Generate datasets
np.random.seed(42)

# 1. Normal distribution (Mean = 0, Std Dev = 1)
normal_data = np.random.normal(0, 1, 1000)

# 2. Right-Skewed (Exponential)
right_skewed_data = np.random.exponential(1, 1000)

# 3. High Kurtosis (Tightly packed normal)
high_kurtosis_data = np.random.normal(0, 0.5, 1000)

# Create a DataFrame
df = pd.DataFrame({
    "Normal": normal_data,
    "Right-Skewed": right_skewed_data,
    "High Kurtosis": high_kurtosis_data
})

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, column in zip(axes, df.columns):
    sns.histplot(df[column], bins=30, kde=True, ax=ax, color='skyblue')
    ax.set_title(f"{column}\nSkewness: {skew(df[column]):.2f}, Kurtosis: {kurtosis(df[column]):.2f}")

plt.tight_layout()
plt.show()
```

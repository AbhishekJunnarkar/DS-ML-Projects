## How to create a Relationship Visualization

```python
# Objective : I want to see the relationship between two continuous variables, wiz, city mileage and co2emmissions
vehicles.plot(kind = 'scatter', x = 'citympg', y = 'co2emissions')
```

```python
# Objective : I want to see the relationship between two continuous variables, wiz, highway mileage and co2emmissions
vehicles.plot(kind = 'scatter', x = 'highwaympg', y = 'co2emissions')
```

```python
# conclusion from the scatter plot: 
# 1. When the emmission is more, the mileage is low
# 2. When the mileage is more, emmission is low
# 3. Majority of vehicles are between 200 to 800 co2emmissions level
# 4. Majority of vehicles mileage in city is between 10 mpg to 30 mpg
```

## How to create a Distribution Visualization

```python
# I want to view the range of CO2 emmissions across the range of given data
vehicles['co2emissions'].plot(kind = 'hist')
```

```python
# conclusion
# Shows that the Co2 emmissions range from 200 grams to 900 grams per mile range
# It also shows that most vehicles emmit 400 to 600 grams per mile range of Co2
```

## How to create a Comparision Visualization

In data science and machine learning, especially using pandas, a pivot refers to restructuring a DataFrame — rearranging rows and columns to make data easier to analyze.

```python
sorted(vehicles['drive'].unique())
```

```python
vehicles.pivot(columns = 'drive', values = 'co2emissions')
```

```python
vehicles.pivot(columns = 'drive', values = 'co2emissions').plot(kind='box',figsize = (10,6))
```

## How to create a Composition Visualization

```python
vehicles.groupby('year')['drive'].value_counts()
```

```python
vehicles.groupby('year')['drive'].value_counts().unstack()
```

```python
vehicles.groupby('year')['drive'].value_counts().unstack().plot( kind = 'bar',stacked = True, figsize = (10,6))
```

## simple contour plot example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
```

```python
# Define a function Z = f(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create the contour plot
plt.contour(X, Y, Z, levels=10, cmap='viridis')
plt.title('Contour Plot of Z = sin(sqrt(X + Y))')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Z value')
plt.grid(True)
plt.show()
```

## Filled Contour Plot
If you want to fill the contours with color:

```python
plt.contourf(X, Y, Z, levels=20, cmap='plasma')
plt.title('Filled Contour Plot')
plt.colorbar()
plt.show()
```

## Add Contour Labels
You can add labels directly on the lines:

```python
contours = plt.contour(X, Y, Z, levels=10, cmap='coolwarm')
plt.clabel(contours, inline=True, fontsize=10)
plt.title('Contour Plot with Labels')
plt.show()
```

## Real-World Use Case: Loss Function in ML
Heres a neat example using a quadratic loss surface often seen in ML optimization:

```python
def loss_fn(w1, w2):
    return w1**2 + 2*w2**2 + 0.5*w1*w2

W1, W2 = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z = loss_fn(W1, W2)

plt.contour(W1, W2, Z, levels=20, cmap='coolwarm')
plt.title('Contour Plot of a Sample Loss Function')
plt.xlabel('w1')
plt.ylabel('w2')
plt.grid(True)
plt.colorbar(label='Loss')
plt.show()
```
